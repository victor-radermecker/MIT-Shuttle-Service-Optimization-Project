```python
tiles = [
            "stamenwatercolor",
            "cartodbpositron",
            "openstreetmap",
            "stamenterrain",
        ]

MIT_Map = folium.Map(
            width=1400,
            height=800,
            location=[42.35676809,-71.09569688],
            zoom_start=15.15,
            tiles=tiles[0],

)
```

```python
residences = pd.read_csv("../HousingData.csv")
```

```python
residences.head()
```

```python
for i in range(len(residences)):
    row = residences.iloc[i]
    print(row)
    folium.Marker(
        location= [row["Lat"], row["Long"]],
        popup= row["Residence"],
        radius=50,
        icon=folium.Icon(color="black", icon="home", prefix="fa"),
    ).add_to(MIT_Map)
```
