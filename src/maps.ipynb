```python
import json
import os
import sys

import branca
import branca.colormap as cm
import geopandas as gpd
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import folium

sys.path.insert(0, "../src")
# from BrusselsMap import BrusselsMap
# from GeoJsonHandler import GeoJsonHandler
# from OSMGeoJsonHandler import OSMGeoJsonHandler

pd.set_option("display.max_rows", 500)
pd.set_option("display.max_columns", 500)

import warnings
warnings.filterwarnings('ignore')
```

```python
tiles = [
            "stamenwatercolor",
            "cartodbpositron",
            "openstreetmap",
            "stamenterrain",
        ]

MIT_Map = folium.Map(
            width=1400,
            height=800,
            location=[42.35676809,-71.09569688],
            zoom_start=15.15,
            tiles=tiles[0],

)
```

```python
residences = pd.read_csv("../HousingData.csv")
current_stops = pd.read_csv("../current_stops.csv")
```

Link for icons: https://fontawesome.com/v4.7/icons/

```python
# Add the residences
for i in range(len(residences)):
    row = residences.iloc[i]
    folium.Marker(
        location= [row["Lat"], row["Long"]],
        popup= row["Residence"],
        radius=50,
        icon=folium.Icon(color="black", icon="home", prefix="fa"),
    ).add_to(MIT_Map)

# Add the bus stops
for i in range(len(current_stops)):
    row = current_stops.iloc[i]
    folium.Marker(
        location= [row["Lat"], row["Long"]],
        popup= row["Station Name"],
        radius=50,
        icon=folium.Icon(color="green", icon="bus", prefix="fa"),
    ).add_to(MIT_Map)

    # Add the path
    if i >= 1:
        point1 = (current_stops.iloc[i]["Lat"], current_stops.iloc[i]["Long"])
        point2 = (current_stops.iloc[i-1]["Lat"], current_stops.iloc[i-1]["Long"])
        folium.PolyLine([point1, point2], color="green", weight=3.5, opacity=1).add_to(MIT_Map)

point1 = (current_stops.iloc[len(current_stops)-1]["Lat"], current_stops.iloc[len(current_stops)-1]["Long"])
point2 = (current_stops.iloc[0]["Lat"], current_stops.iloc[0]["Long"])
folium.PolyLine([point1, point2], color="green", weight=3.5, opacity=1).add_to(MIT_Map)
```
